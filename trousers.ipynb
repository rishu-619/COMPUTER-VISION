```python
from selenium import webdriver
from bs4 import BeautifulSoup
import random
import os
import urllib
import urllib.request

browser=webdriver.Chrome("C:\Program Files (x86)\chromedriver.exe")
browser.get('https://www.amazon.in/s?k=trousers+for+men&page=5')

# soup=BeautifulSoup(browser.page_source)
html = browser.page_source
soup = BeautifulSoup(html)


#do something useful
#prints all the links with corresponding text
all_images = []
for link in soup.find_all('div',{"data-component-type":"s-search-result"},limit=1000):
    href = link.img['src']
    print(href)
    
    img_name = random.randrange(3000,4000)
    
    full_name = str(img_name) + ".jpg"
    
    urllib.request.urlretrieve(href, full_name)
    
    print("loop break")
       
# .get('src'),link.get_text()
```
